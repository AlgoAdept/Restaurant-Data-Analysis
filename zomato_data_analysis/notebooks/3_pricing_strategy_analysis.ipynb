```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import plotly.io as pio
pio.renderers.default = 'notebook'  

# Load cleaned dataset
df = pd.read_csv("../data/zomato_cleaned.csv")
df.head()
```

```python
price_dist = df['Price range'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=price_dist.index, y=price_dist.values, palette='pastel')
plt.title("Distribution of Restaurants by Price Range")
plt.xlabel("Price Range (1 = Cheapest, 4 = Costliest)")
plt.ylabel("Number of Restaurants")
plt.xticks([0,1,2,3], labels=['1','2','3','4'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='muted')
plt.title("Average Ratings by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
delivery_price = df[df['Has Online delivery'] == 'yes']['Price range'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=delivery_price.index, y=delivery_price.values, palette='Blues')
plt.title("Restaurants with Online Delivery by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Count with Delivery")
plt.show()
```

```python
booking_price = df[df['Has Table booking'] == 'yes']['Price range'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=booking_price.index, y=booking_price.values, palette='Greens')
plt.title("Restaurants with Table Booking by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Count with Table Booking")
plt.show()
```

## ðŸ“Œ Insights
- Most restaurants fall into mid-range price categories.
- Higher-priced restaurants receive better ratings on average.
- Delivery is more common among cheaper restaurants.
- Table booking is a premium feature found in high-end places.
